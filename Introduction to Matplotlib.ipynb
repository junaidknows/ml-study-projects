### Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#2. Setup plot # we will use the object oriented method
fig, ax = plt.subplots(figsize=(10, 10)) # Width, height #it's the dimensions of the plot figzie

#3. Plot data
ax.plot(x, y)

#4. Customize plot
ax.set(title="Simple Plot", 
       xlabel="x-axis",
         ylabel="y-axis")

#5. Save & show (you save the whole figure)
fig.savefig("sample-plot1.png")
```

```python
import numpy as np
```

```python
#create some data
x = np.linspace(0, 10, 100)
x[:10] # First 10 elements of x
```

```python
# Plot the data and create a line plot and it's a default plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
plt.show()
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x)) # np.sin(x) is the sine of x
plt.show()
```

```python
# Make a plot from dictionary of Python not from numpy array
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fix, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Store", ylabel="Price ($)")
plt.show()
```

## Horizontal bar plot

```python
#Created a horizontal bar plot using the same data from a python dictionary

nut_butter_prices = {"Almond butter": 10,
                        "Peanut butter": 8,
                        "Cashew butter": 12}
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values())) #ax.barh() is used to create a horizontal bar plot and list is used to convert the dictionary keys and values into a list
ax.set(title="Nut Butter Store", xlabel="Price ($)")
plt.show()
```

```python
# Create some data for histograms and plot it
x = np.random.randn(1000) # randn is used to generate random numbers
fig, ax = plt.subplots() # Create a figure and axes
ax.hist(x) # Plot a histogram
plt.show()
```

## Two options for subplots

```python
# Subplot option 1 , we have tuple in our subplot
# nrows and ncols and figsize are the keyword arguments
# nrows is the number of rows of subplots
# ncols is the number of columns of subplots
# figsize is the dimensions of the plot


fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))


# Commneted out the code because it will not work in the same cell, we're adding the data to each plot below
#plt.show()

# Plot to each different axis
ax1.plot(x, x/2) # Line plot
ax1.set(title="Line Plot")

# Add spacing between plots
plt.tight_layout()

# Plot to 2nd axis
ax2.scatter(np.random.random(10), np.random.random(10)) # Scatter plot
ax2.set(title="Scatter Plot")

# Add spacing between plots
plt.tight_layout()

# Plot to 3rd axis
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values()) # Bar plot
ax3.set(title="Bar Plot")

# Add spacing between plots
plt.tight_layout()

# Plot to 4th axis
ax4.hist(np.random.randn(1000)) # Histogram
ax4.set(title="Histogram")
plt.show()
```

### Subploting Option 2

```python
# Subplot option 2, this is kind of list index way and option 1 was tuple way

fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                          figsize=(10, 5))

# Plot to each different index
ax[0,0].plot(x, x/2) # Line plot
ax[0,0].set(title="Line Plot")

#Add spacing between plots
plt.tight_layout()

ax[0,1].scatter(np.random.random(10), np.random.random(10)) # Scatter plot
ax[0,1].set(title="Scatter Plot")

# Add spacing between plots
plt.tight_layout()

ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values()) # Bar plot
ax[1,0].set(title="Bar Plot")

# Add spacing between plots
plt.tight_layout()

ax[1,1].hist(np.random.randn(1000)) # Histogram
ax[1,1].set(title="Histogram")

plt.show()
```

### Which one to use? (pyplot vs matplotlib Object oriented method)

- When plotting something quickly, okay to use the pyplot method
- When plotting something more advanced, use the OO method

```python
# Object oriented method
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#Pyplot method, plotting directly from the pandas dataframe
over_50.plot(kind="scatter",
                x="age",
                y="chol",
                c="target",
                figsize=(10, 6))
plt.show()
```

```python
# OO Method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
                x="age",
                y="chol",
                c="target",
                ax=ax)

# To adjust the limitation of the x-axis
#ax.set_xlim([50, 80]) # x-axis limit from 50 to 80
plt.show()
```

```python
# OO Method from scratch
fig , ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                        y=over_50["chol"],
                        c=over_50["target"]) # c is used to color the target column values

# Customizing the plot
ax.set(title="Heart Disease and Cholesterol Levels",
         xlabel="Age",
         ylabel="Cholesterol")

# Adding a legend
ax.legend(*scatter.legend_elements(), title="Target") # *scatter.legend_elements() is used to create a legend ``

# Adding a horizontal line
ax.axhline(over_50["chol"].mean(),
            linestyle="--") # Mean of the cholesterol column # it's a keyword

plt.show()
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                                ncols=1,
                                figsize=(10, 10),
                                sharex=True) # sharex is used to share the x-axis, basically it's keeps our plots clean with one name on x-axis rather than display for example age on both plots

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                        y=over_50["chol"],
                        c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
         ylabel="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                        y=over_50["thalach"],
                        c=over_50["target"])
# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
         ylabel="Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a mean line
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")


plt.show()
```
