```python
%matplotlib inline 
# Using magic command, so it will appear in the notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Simplest way to create a plot in Matplotlib
plt.plot()
plt.show()
```

```python
plt.plot([1,2,3,4])
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y)
plt.show()
```

```python
#1st way to create a figure and axes
fig = plt.figure() # Creates a figure
ax = fig.add_subplot() # Adds some axes
plt.show()
```

```python
#2ns way to create a figure and axes
fig = plt.figure() # creates a figure
ax = fig.add_axes([1,1,1,1]) # Adds some axes
ax.plot(x, y) # Adds some data
plt.show()
```

```python
#3rd way to create a figure and axes (recommended)
fig, ax = plt.subplots()
ax.plot(x, y) # Adds some data
plt.show()
```

```python
# Another example of the 3rd method
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 250]) # Plotting data
plt.show()
```

```python
#Object oriented interface
fig, ax = plt.subplots()
ax.plot(x, y) # Adds some data
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")
plt.show()
type(fig), type(ax)

#Type shows that fig is a figure and ax is an axes which is object oriented interface of matplotlib library
```

### Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#2. Setup plot # we will use the object oriented method
fig, ax = plt.subplots(figsize=(10, 10)) # Width, height #it's the dimensions of the plot figzie

#3. Plot data
ax.plot(x, y)

#4. Customize plot
ax.set(title="Simple Plot", 
       xlabel="x-axis",
         ylabel="y-axis")

#5. Save & show (you save the whole figure)
fig.savefig("sample-plot1.png")
```
